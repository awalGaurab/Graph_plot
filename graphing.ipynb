# Data Visualisation with matplotlib

```python
# Import pandas and matplolib libraries
import pandas as pd
from matplotlib import pyplot as plt
```

```python
# read project2 excel file as assign to new dataframe
new = pd.read_excel("project2.xlsx")
print(new.head()) # print few lines of a dataframe
```

Plot a graph of temperature value

```python
plt.plot(new['Date'],new['Temp(C)']) # this statement plot line plot of temperature value.
plt.xlabel("Datetime") # Add label to the x-axis.
plt.ylabel('Temperature') # Add label to the y-axis.
plt.title("Temperature plot") # Add title of the plot.
plt.grid() # Add grid lines to the plot.
plt.legend(['Temperature(°C)']) # Add label of the plotted parameter 
plt.show() # Display plot
```


We can display multiple graphs in a single plot

```python
plt.subplot(2,2,1)
plt.plot(new['Date'],new['Temp(C)'],'g-')
plt.xlabel('Date')
plt.ylabel('Temperature')
plt.xticks(rotation=25)

plt.subplot(2,2,2)
plt.plot(new['Date'],new['Rainfall(mm)'],'r--')
plt.xlabel('Date')
plt.ylabel('Rainfall')
plt.xticks(rotation=25)

plt.subplot(2,2,3)
plt.plot(new['Date'],new['Daylight hours(hr)'],'b')
plt.xlabel('Date')
plt.ylabel('Daylight hours')
plt.xticks(rotation=25)

plt.subplot(2,2,4)
plt.plot(new['Date'],new['windspeed(m/s)'],'c-.')
plt.xlabel('Date')
plt.ylabel('Windspeed')
plt.xticks(rotation=25)

plt.tight_layout(pad=5)
plt.figure(figsize=(10,15))
plt.show()
```
